# Data Exploration

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
unique_job_titles_count = df['job_title'].nunique()
```

```python
unique_job_titles_count
```

```python
top_15_jobs = df['job_title'].value_counts().head(15)
top_15_jobs
```

```python
top_15_jobs.plot(kind= 'bar', color= 'pink') 
plt.title('Top 15 jobs')
plt.ylabel('Number of positions') 
plt.show()
```

```python
maximum_salary = df.groupby('job_title')['salary'].mean().sort_values(ascending = False).head(5) 
maximum_salary.plot(kind = 'bar', width = 0.5, color = 'pink') 
plt.title ('Job with highest salary')
plt.ylabel('Average salary')
plt.show()
```

```python
avg_salary_by_exp_level = df.groupby('experience_level')['salary'].mean() 
plt.figure(figsize=(12, 6))
bars = avg_salary_by_exp_level.plot(kind='bar', color='pink')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Experience Level')
plt.xticks(rotation=45, ha='right')
```

```python
chosen_variable1 = 'salary_in_usd'
chosen_variable2 = 'work_year'

# Plotting histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1) 
plt.hist(df[chosen_variable1], bins=5, color='pink')   #bins is number of interval 
plt.xlabel(chosen_variable1.replace('_', ' ').title())
plt.ylabel('Frequency')
plt.title('Histogram of ' + chosen_variable1.replace('_', ' ').title())
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot(df[chosen_variable2], vert=False, widths=0.5, boxprops=dict(linewidth=2))
plt.xlabel(chosen_variable2.replace('_', ' ').title())
plt.title('Box Plot of ' + chosen_variable2.replace('_', ' ').title())
```
